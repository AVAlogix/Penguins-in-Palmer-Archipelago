```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('penguins_size.csv')

# Display the first few rows of the dataset to understand its structure
print(data.head())

# Checking for missing values
print(data.isnull().sum())
```

```python
# Dropping rows with missing values to clean the dataset
data_cleaned = data.dropna()

# Verifying the cleaned dataset
print(data_cleaned.isnull().sum())
print(data_cleaned.head())
```

```python
# Creating correlation heatmap
plt.figure(figsize=(10, 8))
numeric_cols = ['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']
correlation_matrix = data_cleaned[numeric_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Penguin Measurements')
plt.tight_layout()
plt.show()
```

```python
# Scatterplot for body mass by species and sex
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data_cleaned, x='flipper_length_mm', y='body_mass_g', 
                hue='species', style='sex', s=100)
plt.title('Body Mass vs Flipper Length by Species and Sex')
plt.show()
```

```python
# Box plot for body mass distribution by species
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_cleaned, x='species', y='body_mass_g')
plt.title('Body Mass Distribution by Species')
plt.show()
```

```python
# Violin plot for culmen length by species
plt.figure(figsize=(10, 6))
sns.violinplot(data=data_cleaned, x='species', y='culmen_length_mm')
plt.title('Culmen Length Distribution by Species')
plt.show()
```

```python
# Pair plot for all numeric variables colored by species
sns.pairplot(data_cleaned, hue='species', vars=numeric_cols)
plt.show()
```

```python
# Correlation heatmap for body characteristics by location
plt.figure(figsize=(10, 8))
correlation_matrix_location = data_cleaned.groupby('island')[numeric_cols].mean().corr()
sns.heatmap(correlation_matrix_location, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Penguin Measurements by Island')
plt.tight_layout()
plt.show()
```

```python
# Scatterplot for body mass by location and sex
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data_cleaned, x='flipper_length_mm', y='body_mass_g', 
                hue='island', style='sex', s=100)
plt.title('Body Mass vs Flipper Length by Island and Sex')
plt.show()
```

```python
# Box plot for body mass distribution by island
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_cleaned, x='island', y='body_mass_g')
plt.title('Body Mass Distribution by Island')
plt.show()
```

```python
# Violin plot for culmen length by island
plt.figure(figsize=(10, 6))
sns.violinplot(data=data_cleaned, x='island', y='culmen_length_mm')
plt.title('Culmen Length Distribution by Island')
plt.show()
```

```python
# Pie chart for penguin species distribution
species_counts = data_cleaned['species'].value_counts()
plt.figure(figsize=(8, 8))
species_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange', 'green'])
plt.title('Penguin Species Distribution')
plt.ylabel('')
plt.show()
```

```python
# Flipper length by location - Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=data_cleaned, x='island', y='flipper_length_mm')
plt.title('Flipper Length Distribution by Island')
plt.xlabel('Island')
plt.ylabel('Flipper Length (mm)')
plt.show()
```

```python
# Violin plots for better distribution visualization
plt.figure(figsize=(12, 6))
sns.violinplot(data=data_cleaned, x='island', y='flipper_length_mm')
plt.title('Flipper Length Distribution by Island (Violin Plot)')
plt.xlabel('Island')
plt.ylabel('Flipper Length (mm)')
plt.show()
```

```python
# Flipper length by species - Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=data_cleaned, x='species', y='flipper_length_mm')
plt.title('Flipper Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=data_cleaned, x='species', y='flipper_length_mm')
plt.title('Flipper Length Distribution by Species (Violin Plot)')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.show()
```
